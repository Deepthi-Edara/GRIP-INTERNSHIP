```
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

```
df.describe()
```

```
df.isnull().any()
```

```
df.drop_duplicates(keep = "first",inplace = True)
```

```
corr = df.corr()
sns.heatmap(corr,annot=True)
```

```
df.hist(figsize=(10,10),bins=50)
plt.show()
```

```
df1 = df[['City','Category', 'State','Sales','Profit']]
# pandas pivot with multiple variables
heatmap1_data = pd.pivot_table(df1,values='Sales', index=["City"], columns='Category')
```

```
#Heatmap of sales in every city by category
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap1_data)
```

```
#Heatmap of sales in every State by category
heatmap2_data = pd.pivot_table(df1,values='Sales', index=["State"], columns='Category')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap2_data)
```

```
#Heatmap of profits in every state by category
heatmap3_data = pd.pivot_table(df1,values='Profit', index=["State"], columns='Category')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap3_data)
```

```
#Heatmap of profit in every city by category
heatmap4_data = pd.pivot_table(df1,values='Profit', index=["City"], columns='Category')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap4_data)
```

```
#Heatmap of profit in every city by segment
heatmap5_data = pd.pivot_table(df,values='Profit', index=["City"], columns='Segment')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap5_data)
```

```
#Heatmap of profits in every state by segment
heatmap6_data = pd.pivot_table(df,values='Profit', index=["State"], columns='Segment')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap6_data)
```

```
#Heatmap of sales in every state by segment
heatmap7_data = pd.pivot_table(df,values='Sales', index=["State"], columns='Segment')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap7_data)
```

```
#Heatmap of sales in every city by segment
heatmap8_data = pd.pivot_table(df,values='Sales', index=["City"], columns='Segment')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap8_data)
```

```
#Heatmap of sales in every city by sub-category
heatmap9_data = pd.pivot_table(df,values='Sales', index=["City"], columns='Sub-Category')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap9_data)
```

```
#Heatmap of sales in every state by sub-category
heatmap10_data = pd.pivot_table(df,values='Sales', index=["State"], columns='Sub-Category')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap10_data)
```

```
#Heatmap of profit in every state by sub-category
heatmap11_data = pd.pivot_table(df,values='Profit', index=["State"], columns='Sub-Category')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap11_data)
```

```
#Heatmap of profit in every city by sub-category
heatmap12_data = pd.pivot_table(df,values='Profit', index=["City"], columns='Sub-Category')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap12_data)
```

```
plt.figure(figsize=(10,8))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```
#Profits and sales per sub-category
plt.figure(figsize=(11,7))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```
